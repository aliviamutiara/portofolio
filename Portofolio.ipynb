```python


import matplotlib.pyplot as plt

# Data untuk grafik
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Membuat grafik
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Menambahkan judul dan label
plt.title('Contoh Grafik Garis')
plt.xlabel('X')
plt.ylabel('Y')

# Menampilkan grafik garis
plt.show()
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Random data
np.random.seed(0)
x = np.random.rand(100) * 10  # promosi
y = 3.5 * x + np.random.randn(100) * 9  # Volume penjualan

# Membuat DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# 10 baris pertama data
print(data.head(10))


# Menghitung korelasi Pearson
correlation = data['x'].corr(data['y'])
print(f"Korelasi Pearson: {correlation:.2f}")


# Regresi linier sederhana
slope, intercept, r_value, p_value, std_err = linregress(data['x'], data['y'])

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value**2:.2f}")
print(f"P-value: {p_value:.2f}")
print(f"Standard Error: {std_err:.2f}")

# Membuat plot
plt.figure(figsize=(10, 6))

# Plot data asli
plt.scatter(data['x'], data['y'], color='purple', label='Data')

# Plot garis regresi
plt.plot(data['x'], slope * data['x'] + intercept, color='blue', linewidth=2, label='Garis Regresi')

# Menambahkan judul dan label
plt.title(' Contoh Regresi Linier Sederhana')
plt.xlabel('Promosi (x)')
plt.ylabel('Volume penjualan (y)')
plt.legend()

# Plot visualization
plt.show()
```
